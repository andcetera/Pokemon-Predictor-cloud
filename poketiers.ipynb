```python
import sqlite3
from pathlib import Path
import pandas as pd
```

```python
database_path = "Resources/pokemon_with_tiers.sqlite"
Path(database_path).touch()
```

```python
conn = sqlite3.connect(database_path)
c = conn.cursor()
```

```python
c.execute('''DROP TABLE IF EXISTS poketiers;''')
```

```python
c.execute('''CREATE TABLE poketiers(
    _ID INT NOT NULL PRIMARY KEY,
    Number INT NOT NULL,
  	Name VARCHAR(40) NOT NULL,
    Type_1 VARCHAR(40),
	Type_2 VARCHAR(40),
	HP INT NOT NULL,
	Attack INT NOT NULL,
	Defense INT NOT NULL,
	Sp_Atk INT NOT NULL,
	Sp_Def INT NOT NULL,
	Speed INT NOT NULL,
	Generation INT NOT NULL,
	Legendary INT NOT NULL,
	Tier VARCHAR(40) NOT NULL
	);''')
```

```python
df = pd.read_csv('Resources/pokemon_with_tiers.csv')
```

```python
df.to_sql("poketiers", conn, if_exists='append', index = False)
```

```python
c.execute('''SELECT * from poketiers''').fetchall()
```

```python
conn.close()
```
